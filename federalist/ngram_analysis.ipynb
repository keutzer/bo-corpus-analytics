```python
from collections import Counter
from nltk.util import ngrams

import pickle
```

```python
with open('bibliography', 'rb') as f:
    bib = pickle.load(f)
```

```python
h_essays = []
m_essays = []
for i in bib['HAMILTON']:
    with open('lower_no_punc/federalist_no_{:02}.txt'.format(i), 'r') as f:
        h_essays.append(f.read().split())
for i in bib['MADISON']:
    with open('lower_no_punc/federalist_no_{:02}.txt'.format(i), 'r') as f:
        m_essays.append(f.read().split())
```

## Explore top 3-grams

```python
N = 3
print("Hamilton essays")
for e in h_essays:
    c = Counter(ngrams(e, N))
    print('\n'.join([' '.join(k) + ': {}'.format(v) for k, v in c.most_common()[:3]]), '\n')
    
print("Madison essays")
for e in m_essays:
    c = Counter(ngrams(e, N))
    print('\n'.join([' '.join(k) + ': {}'.format(v) for k, v in c.most_common()[:3]]), '\n')
```
